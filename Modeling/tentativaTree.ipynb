```python


from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.preprocessing import LabelEncoder
import dtreeviz 
import pandas as pd
import numpy as np


df = pd.read_csv("/Users/carolbarroco/Documents/RecomendSisKNN/TG2 RecommendSys/GitHub/T2-Indroducao-Ciencia-de-Dados/Data/MPBROCKMETAL_KGDataSet.csv")

X = df.iloc[:, 6:20]
y = df["track_genre"]
le = LabelEncoder()
y_encoded = le.fit_transform(y)
feature_names = X.columns

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.3, random_state=42)

# Retrain model
dtree_clf = DecisionTreeClassifier(max_depth=5)
dtree_clf.fit(X_train, y_train)

# Predict on test set
y_pred_test = dtree_clf.predict(X_test)

# viz_model = dtreeviz.model(
#     dtree_clf,
#     X_train=X,
#     y_train =y_encoded,
#     feature_names=feature_names,
#     target_name="track_genre",
#     class_names=list(y.unique())
# ) 

viz_model = dtreeviz.model(
    dtree_clf,
    X_train=X_test,
    y_train =y_test,
    feature_names=feature_names,
    target_name="track_genre",
    class_names=list(le.classes_)
) 


# Show the visualization (works in Jupyter or saves as SVG)
viz_model.view()  # or .save("tree.svg")
```

```python
# Pick a sample (row) to analyze
sample_index = 2  # change this to inspect other tracks
sample = X_test.iloc[[sample_index]]  # double brackets to keep as DataFrame

# Get the path of nodes this sample passed through
node_indicator = dtree_clf.decision_path(sample)

# Get the node feature splits
tree = dtree_clf.tree_
features = tree.feature
thresholds = tree.threshold

print(f"\nðŸŽ§ Decision path for sample {sample_index} (true genre: {int(y_test[sample_index])})")

for node_id in node_indicator.indices:
    if features[node_id] != -2:  # -2 means it's a leaf node
        feature_name = feature_names[features[node_id]]
        threshold = thresholds[node_id]
        sample_value = sample.iloc[0, features[node_id]]

        direction = "â‰¤" if sample_value <= threshold else ">"
        print(f"Node {node_id}: {feature_name} {direction} {threshold:.4f} (sample value: {sample_value:.4f})")
    else:
        print(f"ðŸŸ© Leaf node {node_id} reached")
```
